```python
from PIL import Image
from IPython.display import display

# Open image
img = Image.open("../images/images.jpg")

# Display image
display(img)
```

```python
import numpy as np

img_array = np.array(img) #image laii numpy array maa convert gareko

print("Shape:", img_array.shape)
print(img_array) # output:rgb number 
```

```python
import matplotlib.pyplot as plt

# Split image into left and right halves
height, width, channels = img_array.shape # height, width, channel(RGB)

left = img_array[:, :width//2] #if w=600 then, w//2= 600/2=300 ie:= take all rows, :300 take columns from 0-299
right = img_array[:, width//2:]

# Join horizontally
horizontal = np.hstack((left, right))

plt.imshow(horizontal)
plt.axis("off")
plt.show()
```

```python
# Stack image vertically
vertical = np.vstack((img_array, img_array))

plt.imshow(vertical)
plt.axis("off")
plt.show()
```

```python
flipped = np.fliplr(img_array)
#flip img horizontally
plt.imshow(flipped)
plt.axis("off")
plt.show()
```

```python
flipped_v = np.flipud(img_array)
#flip img vertically top-bottom
plt.imshow(flipped_v)
plt.axis("off")
plt.show()
```

```python
cropped = img_array[50:150, 80:200]
#crop image specific part
plt.imshow(cropped)
plt.axis("off")
plt.show()
```

```python
rotated = np.rot90(img_array)
# rotate image 90 degree
plt.imshow(rotated)
plt.axis("off")
plt.show()
```

```python
dark = img_array * 0.5
dark = dark.astype(np.uint8)
# reduce brightness (darken image)
plt.imshow(dark)
plt.axis("off")
plt.show()
```

```python
bright = img_array * 1.5
bright = np.clip(bright, 0, 255).astype(np.uint8)
#brighten image
plt.imshow(bright)
plt.axis("off")
plt.show()
```

```python
import cv2
#blur img
blur = cv2.GaussianBlur(img_array, (7,7), 0)

plt.imshow(blur)
plt.axis("off")
plt.show()
```

```python
red_filter = img_array.copy()
red_filter[:, :, 1] = 0  # remove green
red_filter[:, :, 2] = 0  # remove blue

plt.imshow(red_filter)
plt.axis("off")
plt.show()
```

```python
stacked = np.hstack((img_array, img_array)) #hstack and vstack

plt.imshow(stacked)
plt.axis("off")
plt.show()
```
